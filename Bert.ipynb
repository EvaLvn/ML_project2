```python
df = pd.read_csv('out_clean.csv')
```

```python
def remove_nan(text):
    return re.sub(r'nan', r' ', str(text))
```

```python
df["clean"] = df["clean"].apply(lambda x: str(x)).apply(lambda x: remove_nan(x))
```

# Topics for deutsch tweets

```python
tweets = df[df.withheld_in_countries == '[\'DE\']'].clean.to_list()
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")
s = list(stopwords.words('english')) + ['http', 'https', 'amp', 'com', 'gt', 'lol', 'true', 'laureen', 'pink']
```

```python
vectorizer_model = CountVectorizer(ngram_range=(1, 2), stop_words=s)

model = BERTopic(
    vectorizer_model=vectorizer_model,
    embedding_model=embedding_model,
    language='english', calculate_probabilities=True,
    verbose=True,
    min_topic_size=7
)
topics, probs = model.fit_transform(tweets)
```
